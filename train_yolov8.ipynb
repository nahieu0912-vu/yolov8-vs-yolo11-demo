This notebook provides a step-by-step process to train a YOLOv8 object detection model using the dataset provided in this project.

## 2. Model Initialization
We will load a pre-trained YOLOv8 nano model (`yolov8n.pt`) as a starting point for our training.

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
model = YOLO('yolov8n.pt')
```

## 3. Model Training
Now we will train the model using the `data.yaml` file. We'll set the number of epochs, image size, and batch size.

```python
# Train the model using GPU

results = model.train(
    data="data.yaml",
    epochs=197,
    imgsz=640,
    batch=16,        # Reduced batch size for a single GPU
    device=1         # Use the first available GPU
           
)
```

## 4. Model Validation
After training, we validate the model on the validation set to check its performance.

```python
# Validate the model
metrics = model.val()
print(metrics)
```

## 5. Model Inference
Finally, we can use the trained model to perform detection on new images.

```python
# Run inference on an image from the test set
# Replace 'test/images/your_image.jpg' with an actual image path from your dataset
results = model.predict(source='test/images', save=True, conf=0.5)

# Display results
for result in results:
    print(result.boxes) # print boxes to console
    result.show() # This will open the image with detections in a window
```

## 6. Testing on Real-World Images
Now we will test the trained model on new images captured by a phone, located in the `real_image` folder.

```python
# Run inference on real-world images
# Ensure the 'real_image' folder exists in your project directory
results_real = model.predict(source='real_image', save=True, conf=0.5)

# Display results for real images
for result in results_real:
    print(result.boxes)
    result.show()
```

## 8. Training with YOLO11s
YOLO11 utilizes an optimized CNN architecture featuring advanced C3k2 feature-extraction modules. The Small ('s') variant provides an ideal balance of processing speed and detection accuracy, delivering better feature retention than the ultra-lightweight Nano ('n') model. Here is how to initialize and train using YOLO11s.

```python
from ultralytics import YOLO


# Load a pretrained YOLO11s model
model_11 = YOLO('yolo11s.pt')

# Train the YOLO11s model
results_11 = model_11.train(
    data="data.yaml",
    epochs=100,       # Adjusted epochs for demonstration 100 instead of 197
    imgsz=640,
    batch=16,
    device=0
)
```

## 9. Testing with YOLO11s
Now we will use the YOLO11s model to perform inference on the test images to evaluate its performance.

```python
# Run inference on the test set using the YOLO11s model
results_11_test = model_11.predict(source='test/images', save=True, conf=0.5)

# Display results
for result in results_11_test:
    print(result.boxes)
    result.show()
```
